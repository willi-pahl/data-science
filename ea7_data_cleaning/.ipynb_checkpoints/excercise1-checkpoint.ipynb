```python
# Data set load.
import numpy as np
import pandas as pd
pathname = '/home/pahl/Dokumente/Studium/6_ds/github/ea7_data_cleaning/'
filename = 'dsm-beuth-edl-demodata-dirty.csv'
df = pd.read_csv(pathname + filename)
df
```

```python
# Empty data set, 18, 23
# Delete all records and/or rows that are empty.
df.dropna(how='all', inplace=True)
df
```

```python
# Double data set, 15, 16
df.drop_duplicates(subset=['full_name','first_name','last_name','email','gender','age'], keep='first', inplace=True)
df
```

```python
# Empty data set id, 22
# 
numbers = list(range(1, df.shape[0] + 3))
daFr = pd.DataFrame({'id': numbers})
df.update(daFr)
df
```

```python
# Age is not integer, 10,
# If the age specification is a word, then it becomes the value zero.
df.loc[df['age'].str.isalpha(), 'age'] = 0
df
```

```python
# Age is negativ integer.
# A negative age is converted to a positive age in hopes that it was a typo.
df['age'] = df['age'].astype(str).astype(int)
df.loc[df['age'] < 0, 'age'] = df['age'].abs()
df
```

```python
# Privacy, remove all records under the age of 16, 4, 11, 17
# For data protection reasons, remove all persons who are not of legal age. Exclude the previously set 0 years.
df.drop(df.loc[(df['age'] < 18) & (df['age'] > 0), 'age'].index, inplace=True)
df
```
