```python
import gym

actions = {
    'Left': 0,
    'Down': 1,
    'Right': 2, 
    'Up': 3
}

print('---- winning sequence ------ ')
winning_sequence = (2 * ['Right']) + (4 * ['Down']) + (2 *['Right'])*10
print(winning_sequence)

custom_map = [
    'SFFHF',
    'FFFFF',
    'FFFFH',
    'FFFFF',
    'HFFFG'
]

env = gym.make("FrozenLake-v1", desc=custom_map)
env.reset()
env.render()

for a in winning_sequence:
    new_state, reward, done, info = env.step(actions[a])
    print()
    env.render()
    print("Reward: {:.2f}".format(reward))
    print(info)
    if done:
        break   

print()
env.close()
```
